```python
response = requests.get('https://query1.finance.yahoo.com/v8/finance/chart/MSFT?start=2020-09-26&end=2020-09-27&interval=1h')
```

```python
response
```

```python
def get_chart(ticker, start, end, interval):
    url = 'https://query1.finance.yahoo.com/v8/finance/chart/'
    return requests.get(f'{url}{ticker}?start={start}&end={end}&interval={interval}').json()
```

```python
get_chart('MSFT', start='2020-09-26', end='2020-09-27', interval='1h')
```

```python
from datetime import datetime, timedelta
```

```python
from_time = int((datetime.now() - timedelta(minutes=1000)).timestamp())
```

```python
response = requests.get(f'https://api.kraken.com/0/public/OHLC?pair=ETHBTC&interval=30&since={from_time}')
r = response.json()['result']['XETHXXBT']
df = pd.DataFrame(r, columns=['time', 'open', 'high', 'low', 'close', 'vwap', 'volume', 'count'])
df['open'] = pd.to_numeric(df['open'])
df['time'] = df['time'].apply(lambda x: datetime.fromtimestamp(x))
df
```

```python
df['open'].plot()
```
